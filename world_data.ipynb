```python
def read_stata_files(base_path):
    res = []
    for (dirpath, dirnames, filenames) in os.walk(base_path):    
        res.extend([f for f in filenames if 'dta' in f and not 'china' in f])
    return res

def read_sas_files(base_path):
    res = []
    for (dirpath, dirnames, filenames) in os.walk(base_path):    
        res.extend([f for f in filenames if 'sas7bdat' in f and not 'china' in f])
    return res
```

```python
df_list = []
for sf in read_sas_files(base_path):
    file_name = base_path + sf
    print '\nhandling', file_name
    print 'reading SAS7BDAT'
    with  sas7bdat.SAS7BDAT(file_name) as f:
        try:
            print 'converting to dataframe'
            df = f.to_data_frame()
            print 'merging year columns'
            df['ValidYear'] = df.apply(lambda r: r['Year'] if pd.notnull(r['Year']) else r['year'])
            print 'assigning ValidYear to Year'
            df['Year'] = df['ValidYear']
            print 'dropping ValidYear'
            df.drop('ValidYear')
            print 'dropping year'
            df.drop('year')
            pickle_file = '.\\' + sf + '.pickle'
            print 'serializing pickle file', pickle_file
            df.to_pickle(pickle_file)
        except Exception as ex:
            print ex
            pass
```
